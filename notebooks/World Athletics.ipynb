```python
import pandas as pd
```

```python
df = pd.read_csv('../data/external/treadmill_incline.csv')
```

```python
def min_per_mile_to_min_per_km(min_per_mile):
    min_per_mile, seconds_per_mile = min_per_mile.split(":")
    pace = (int(min_per_mile) + int(seconds_per_mile) / 60) / 1.609344
    min_per_km = int(pace)
    seconds_per_km = round(pace % 1 * 60)
    return f"{min_per_km}:{seconds_per_km:02.0f}"


df["min/km"] = df["Pace per mile"].apply(min_per_mile_to_min_per_km)
df["Treadmill Km/h"] = (df["Treadmill MPH setting"] * 1.609344).round(2)

df = df.drop(columns=["Pace per mile", "Treadmill MPH setting"])

columns = [f"{i}%" for i in range(11)]
df.loc[:, columns] = df.loc[:, columns].applymap(min_per_mile_to_min_per_km)

df = df.set_index("Treadmill Km/h")
df = df[[df.columns[-1]] + list(df.columns[:-1])]
df.index = df.index.to_series().apply("{:.02f} km/h".format)
```

```python
df.index = df.reset_index()["Treadmill Km/h"].apply(lambda x: x.split(" ")[0])
df.index = df.index.astype(float)
df = df.drop(columns=["min/km"])
```

```python
def min_per_km_to_pace(min_per_km):
    min_per_km, seconds_per_km = min_per_km.split(":")
    pace = int(min_per_km) + int(seconds_per_km) / 60
    return pace

columns = [f"{i}%" for i in range(11)]
df.loc[:, columns] = df.loc[:, columns].applymap(min_per_km_to_pace)
```

```python
tmp = df.reset_index()
```

```python
X, y = [], []
for clm in tmp.columns[2:]:
    clm_value = int(clm.replace("%", ""))
    for i, val in enumerate(tmp[clm].values):
        # X.append([clm_value, val])
        X.append([clm_value, 1 / val])
        y.append(tmp.loc[i, "Treadmill Km/h"])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor()
model.fit(X, y)
```

```python
grade = 0
pace = 7.78333333333333

treadmill_kmh = model.predict([[grade, pace]])
treadmill_kmh = treadmill_kmh[0]

treadmill_kmh
```
