```python
firstname = 'hugo'
lastname = 'poussier'
# licence_number = '1174307'
licence_number = ''
```

```python
results = pd.DataFrame()
for year in range(2008, 2024):
    print(year)
    try:
        url = f"https://bases.athle.fr/asp.net/liste.aspx?frmpostback=true&frmbase=resultats&frmmode=1&frmespace=0&frmsaison={year}&frmclub=&frmnom={lastname}&frmprenom={firstname}&frmsexe=&frmlicence={licence_number}&frmdepartement=&frmligue=&frmcomprch="
        df = pd.read_html(url, skiprows=2, attrs={'id': 'ctnResultats'})[0]
        df[0] = pd.to_datetime(df[0].astype(str) + f"/{year}", format="%d/%m/%Y")
        df = df.drop(columns=[1, 3, 5, 7, 9, 11, 13, 15, 17])
        df.columns = ['Date', 'Competitor', 'Discipline', 'Serie', 'Place', 'Result', 'Score', 'Category', 'Performance', 'Venue',]
        results = pd.concat([results, df])
    except ValueError:
        pass
    
```

```python
results
```

```python
# give me a slice of the results dataframe with only the 800m events (discipline has 800m in the name)
perf_800m = results[results.Discipline.str.contains('800m')].copy()

# transform the result column into seconds (result in format mm'ss"cc") but it can be DNF or NP
perf_800m['Result'] = perf_800m['Result'].apply(lambda x: 0 if x in ['NP', 'DNF'] else int(x.split("'")[0])*60 + int(x.split("'")[1].split('"')[0]))
```

```python
perf_800m
```

```python
perf_800m[perf_800m.Result != 0.0].plot(x='Date', y='Result', kind='scatter')
```
