```python
import pandas as pd
```

```python
url = "https://bases.athle.fr/asp.net/liste.aspx?frmbase=resultats&frmmode=1&frmespace=0&frmcompetition=283234&frmepreuve=13+-+Course+Elite+H+(ES-SE-M0)+TCM"
df = pd.read_html(url, skiprows=4, attrs={"id": "ctnResultats"})[0]
df = df.drop(columns=[1, 3, 5, 7, 9, 11, 13, 14, 15, 16])
df.columns = [
    "Place",
    "Temps",
    "Name",
    "Club",
    "Département",
    "Région",
    "Categorie",
]
df["Temps"] = df.Temps.apply(lambda row: row.split(" ")[0])
df["Nom"] = df["Name"].apply(
    lambda row: "".join([l for l in row if l.isupper() or l == " "][:-1])
)
df["Prénom"] = df.apply(lambda row: row.Name.split(row.Nom)[1], axis=1)
df["Nom"] = df.Nom.str.strip()
df = df.set_index("Place")
df = df.drop(columns=["Name", "Département", "Région"])
```

```python
rps = []
for idx, row in df.iterrows():
    print(row.Nom, row.Prénom)
    year = 2024
    found_rp = False
    while year > 2015 and not found_rp:
        url = f"https://bases.athle.fr/asp.net/liste.aspx?frmpostback=true&frmbase=bilans&frmmode=2&frmespace=0&frmannee={year}&frmepreuve=261&frmnom={row.Nom}&frmprenom={row.Prénom}&frmlicence=&frmclub=&frmligue=&frmdepartement="
        try:
            tmp = pd.read_html(url, skiprows=2)[0]
            rp = tmp[6]
        except:
            year -= 1
            continue
        for r in rp:
            if "(RP)" in r:
                found_rp = r.split(" ")[0]
                rps.append(found_rp)
                print(found_rp)
                break
        year -= 1
    if not found_rp:
        rps.append("")
```

```python
df['Record 10 km'] = rps
```
